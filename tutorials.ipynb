```python
# Variables / Ozgaruvchilar

# Ozgaruvchi yaratish sintaksisi
# variable_name = value

# variable_name / ozgaruvchining nomi / ozgaruvchi
# = / ozlashtirish amali
# value / qiymati

# ozgaruvchi nomida ishlatilishi mumkin bolgan belgilar royxati
# variable_name: [A-Z, a-z, 0-9, _]
# ozgaruvchi nomi boshida raqamlarni ishlatib bolmaydi

domain_name = "cyberspace.uz"

print(f"Domain Name: {domain_name}.")
```

```python
# Data Types / Malumot turlari

# int(integer) / butun son
a = 7

# output variable and data type
print(a, type(a))
```

```python
# float(double) / haqiqiy son

b = 2.7

print(b, type(b))
```

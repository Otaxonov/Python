```python
# if statement

# if condition:
#    code
```

```python
a, b = 14, 5

if a > b: # True
    print(True)
```

```python
# if..else statement

# if condition:
#   code
# else:
#   code

a, b = 7, 15

if a > b: # True
    print(a)
else: # False
    print(b)
```

```python
# Nested if..else
a, b, c = 1, 2, 3

if a > b:
    if a > c:
        print(f"Max: {a}")
    else:
        print(f"Max: {c}")
else:
    if b > c:
        print(f"Max: {b}")
    else:
        print(f"Max: {c}")
```

```python
# if..elif..else statement
a, b = 4, 4

if a > b:
    print(f"{a} is greater than {b}")
elif a < b:
    print(f"{a} is less than {b}")
else:
    print(f"{a} is equal to {b}")
```

```python
age = 25

if age < 13:
    print("You are a child")
elif age < 20:
    print("You are a teenager")
elif age < 65:
    print("You are a adult")
elif age >= 65:
    print("You are an elder")
```

```python
# Loops / Sikl
"""
1. for
2. while
"""

# [1, 10]
# (1, 10)
# (1, 10]
# [1, 10)

# range(start, stop, step)

# range(10) -> range(0, 10, 1)
for i in range(10):
    print(i, end=" ")
```

```python
# range(1, 11) -> range(1, 11, 1)
for i in range(1, 11):
    print(i, end=" ")
```

```python
for i in range(1, 11, 2):
    print(i, end=" ")
```

```python
# While Loop

i = 0
while i <= 10:
    print(i, end=" ")
    i += 1
```

```python
# Input Method/Function / Malumotni ekrandan oqib olish metodi
name = input("What is your name?:")

print(f"Hello, {name}!")    # f-string
```

```python
# Variables / Ozgaruvchilar

# Ozgaruvchi yaratish sintaksisi
# variable_name = value

# variable_name / ozgaruvchining nomi / ozgaruvchi
# = / ozlashtirish amali
# value / qiymati

# ozgaruvchi nomida ishlatilishi mumkin bolgan belgilar royxati
# variable_name: [A-Z, a-z, 0-9, _]
# ozgaruvchi nomi boshida raqamlarni ishlatib bolmaydi

domain_name = "cyberspace.uz"

print(f"Domain Name: {domain_name}.")
```

```python
# Data Types / Malumot turlari

# int(integer) / butun son
a = 7

# output variable and data type
print(a, type(a))
```

```python
# float(double) / haqiqiy son

b = 2.7

print(b, type(b))
```

```python
# str(string) / satr

first_name = "Sarvar"
last_name = 'Otaxonov'

print(last_name, first_name)

print(type(last_name), type(first_name))
```

```python
# bool(boolean) / faqat ikki qiymat(rost/yolgon, true/false, 1/0) qabul qiluvchi malumot turi

is_admin = True
print(is_admin, type(is_admin))

# n = 5
is_odd = True
is_even = False
```

```python
# Arithmetic operators / Arifmetik amallar [+, -, *, /, %, //, **]
a = 5
b = 7

c = a + b
print(f"{a} + {b} = {c}")
```

```python
# kop ozgaruvchilarni bir qatorda ozlashtirish sintaksisi

a, b = 5, 7
```

```python
# Type Casting / Bir malumot turidan ikkinchi malumot turiga otish(convert)
# int(), float(), bool(), str()

a = "12"
b = "13"
print(a + b)
print(a, type(a))
```

```python
# str to int / satr malumot turidan butun son tipiga otish

b = int(a)

print(b, type(b))
```

```python
# Assignment operators / Ozlashtirish amallari
# [ =, +=, -=, *=, /=, %=, //=, **= ]

a = 7
print(a)
```

```python
a, b, c = 1, 2, 3
print(a, b, c)
```

```python
# Data Types / Malumot turlari

# List / royhat
list1 = [3, 2.7, True, 'Bobur', "Qadamov"]

print(list1)
```

```python
# Tuple / kartej
tuple1 = (3, 2.7, True, 'Bobur', "Qadamov")
print(tuple1)
```

```python
# Set / Toplam
set1 = {1, 1, 1, 3, 3, 3, 4, 4, 4}

print(set1)
```

```python
# Dictionary / lugat
# dict = { "key": "value" }

student = {
    "firstname": "Bobur",
    "lastname": "Qadamov",
}

print(student)
```

```python
first_name = student.get("firstname")

print(first_name)
```

```python
first_name = student['firstname']

print(first_name)
```
